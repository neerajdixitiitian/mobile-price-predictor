```python
import time
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
```

```python
def wait_for_page_to_load(driver,wait):
    title = driver.title
    try:
        wait.until(
            lambda d: d.execute_script("return document.readyState") == "complete"
        )
    except:
        print(f"The webpage \"{title}\" did not get fully loaded.\n")
    else:
        print(f"The webpage \"{title}\" did get fully loaded.\n")

# options
chrome_options = Options()
chrome_options.add_argument("--disable-http2")
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("--ignore-certificate-errors")
chrome_options.add_argument("--enable-features=NetworkServiceInProcess")
chrome_options.add_argument("--disable-features=NetworkService")
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36"
)

driver = webdriver.Chrome(options = chrome_options)
driver.maximize_window()

# explicit wait
wait = WebDriverWait(driver,10)

# accessing to web page 
url = "https://www.flipkart.com/"
driver.get(url)
wait_for_page_to_load(driver,wait)

# entering the text to search bar
try:
    search_bar = wait.until(
        EC.presence_of_element_located((By.XPATH,'//*[@id="container"]/div/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div/header/div[1]/div[2]/form/div/div/input'))
    )
except:
    print("Timeout while locating the webpage")
else:
    search_bar.send_keys("android mobile 5g")
    time.sleep(2)

# click on right option
try:
    right_option = wait.until(
        EC.element_to_be_clickable((By.XPATH,'//*[@id="container"]/div/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div/header/div[1]/div[2]/form/ul/li[2]/div'))
    )
except:
    print("Timeout while clicking right option")
else:
    right_option.click()
    time.sleep(3)
# apply only flipkart assured
try:
    assured_button = wait.until(
        EC.element_to_be_clickable((By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[1]/div/div[1]/div/section[4]/label/div[2]/div/img'))
    )
except:
    print("Timeout while clicking on button")
else:
    assured_button.click()
    time.sleep(2)
    
# scrolling the webpage
data = [] 
page_count = 0
while True:
    page_count += 1 
    print(f"scraping page {page_count}")
    try:
        next_page_button = driver.find_element(By.XPATH,"//span[normalize-space()='Next']")
    except:
            print(f"Timeout because we have navigated through all {page_count} pages.\n")
            break
    else:
        try:
            driver.execute_script("window.scrollBy(0,arguments[0].getBoundingClientRect().top - 150);", next_page_button)
            time.sleep(2)
            # scrape the data
            rows = driver.find_elements(By.CLASS_NAME,"_75nlfW")
            for row in rows:
                # mobile name
                try:
                    name = row.find_element(By.CLASS_NAME, "KzDlHZ").text
                except:
                    name = np.nan
                # mobile price
                try:
                    price = row.find_element(By.CLASS_NAME,"Nx9bqj").text
                except:
                    price = np.nan
                # mobile rating
                try:
                    rating = row.find_element(By.CLASS_NAME,"XQDdHH").text
                except:
                    rating = np.nan
                # scrape elements
                try:
                    specs = row.find_elements(By.CSS_SELECTOR,".J\\+igdf")
                    ram = specs[0].text if len(specs) > 0 else np.nan
                    display = specs[1].text if len(specs) > 1 else np.nan
                    camera = specs[2].text if len(specs) > 2 else np.nan
                    battery = specs[3].text if len(specs) > 3 else np.nan
                    processor = specs[4].text if len(specs) > 4 else np.nan
                except:
                    ram = display = camera = battery = processor = np.nan
                data.append({
                    "name":name,
                    "price":price,
                    "rating":rating,
                    "ram":ram,
                    "display":display,
                    "camera":camera,
                    "battery":battery,
                    "processor":processor
                })
                           
            wait.until(
                EC.element_to_be_clickable((By.XPATH,"//span[normalize-space()='Next']"))
            ).click()
            time.sleep(5)
        except:
            print("Timeout while clicking the \"Next Page\".\n")

# scrolling and scraping the last page 
# scrape the data
rows = driver.find_elements(By.CLASS_NAME,"_75nlfW")
for row in rows:
    # mobile name
    try:
        name = row.find_element(By.CLASS_NAME, "KzDlHZ").text
    except:
        name = np.nan
    # mobile price
    try:
        price = row.find_element(By.CLASS_NAME,"Nx9bqj").text
    except:
        price = np.nan
    # mobile rating
    try:
        rating = row.find_element(By.CLASS_NAME,"XQDdHH").text
    except:
        rating = np.nan
    # scrape elements
    try:
        specs = row.find_elements(By.CSS_SELECTOR,".J\\+igdf")
        ram = specs[0].text if len(specs) > 0 else np.nan
        display = specs[1].text if len(specs) > 1 else np.nan
        camera = specs[2].text if len(specs) > 2 else np.nan
        battery = specs[3].text if len(specs) > 3 else np.nan
        processor = specs[4].text if len(specs) > 4 else np.nan
    except:
        ram = display = camera = battery = processor = np.nan
    data.append({
        "name":name,
        "price":price,
        "rating":rating,
        "ram":ram,
        "display":display,
        "camera":camera,
        "battery":battery,
        "processor":processor
    })
time.sleep(2)
driver.quit()
```
